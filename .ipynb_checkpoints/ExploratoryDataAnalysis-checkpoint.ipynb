# Numerical variable

```python
def num_feature(dataset):
    
    """
    This function takes a dataset as input and returns four lists of features: 
    year_features, numerical_features, discrete_feature, and continuous_feature.
    
    year_features: a list of date variables in the dataset
    numerical_features: a list of numerical variables that are not date variables or Id column
    discrete_feature: a list of discrete variables that have a limited number of unique values
    continuous_feature: a list of continuous variables that are not date variables or discrete variables
    """
        
    # list of numerical variables
    year_features = [features for features in dataset.columns if dataset[features].dtypes == 'datetime64[ns]']
    
    numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'object' and feature not in year_features+['Id']]
    

    ## Numerical variables are usually of 2 type
    ## 1.Discrete Variables
    discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_features+['Id']]
    
    ## 2.Continuous variable
    continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_features+['Id']]
   
    return year_features, numerical_features, discrete_feature, continuous_feature
```

```python
year_features, numerical_features, discrete_feature, continuous_feature = num_feature(client_df)
```

```python
print('Number of year variables: ', len(year_features))
year_features
```

```python
print('Number of discrete variables: ', len(discrete_feature))
discrete_feature
```

```python
print('Number of continuous variables: ', len(continuous_feature))
continuous_feature
```

```python
def describe_variables(dataset, variables):
    
    """
    Prints the number of unique values and value counts for each feature in the variables list.

    Args:
    dataset: pandas DataFrame
        The dataset to be analyzed.
    variables: list
        The list of features to be analyzed.

    Returns:
    None
    """
    
    for feature in variables:
        print(f'Column: {feature}')
        print(f'Number of unique values: {dataset[feature].nunique()}')
        describe_frame = dataset[[feature]].value_counts().reset_index(name='count')
        print(describe_frame,'\n'+'--'*40)  
```

```python
describe_variables(client_df, numerical_features)
```

This feels very cluttered, instead we can pass our desired variables to look at.

```python
def plot_countplot_with_percentage(data, feature, val = 0):
    total = len(data[feature])
    ax = sns.countplot(x=feature, hue='churn', data=data)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width()/2
        y = p.get_y() + p.get_height() + 0.02
        ax.annotate(percentage, (x, y), ha='center')
    # Rotate xticks
    ax.set_xticklabels(ax.get_xticklabels(), rotation=val)
    plt.show()
```

```python
plot_countplot_with_percentage(client_df, 'churn')
```

* This is an imbalance dataset.
* Around 9.7% of clients have churned.

```python
## Lets Find the realtionship between discrete variables and churn

for feature in discrete_feature:
    data_client=client_df.copy()
    sns.countplot(x=feature, hue='churn', data=data_client)
    plt.xlabel(feature)
    plt.ylabel('churn')
    plt.title(feature)
    plt.show()
```

```python
# Create countplots with a logarithmic y-axis scale
for feature in discrete_feature:
    data_client = client_df.copy()
    ax = sns.countplot(x=feature, hue='churn', data=data_client)
    ax.set_yscale('log')
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature + ' (log scale)')
    plt.show()
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data_client=client_df.copy()
    sns.histplot(x = feature, data = data_client)
    #data_client[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

* The numeric data is not at all guassian distributed rather somewhat right skewed.
* We can apply log transformation to see how the data looks.
* We can see there are alot of outliers in the data.

```python
# We will be using logarithmic transformation


for feature in continuous_feature:
    data_client=client_df.copy()
    data_client[feature]=np.log1p(data_client[feature])
    sns.histplot(x = feature, data = data_client)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
        
```

* We can see that log transformation gives a nice normal curve for most of the values.
* We can see a consolidation of values at 0 for almost all features.

```python
## Let us take a look at outlies

for feature in continuous_feature:
    data_client=client_df.copy()
    data_client[feature]=np.log1p(data_client[feature])
    #sns.boxplot(x = data_client[feature], orient = 'h')
    data_client.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

**We can see a lot of outliers in the data**

```python
#Spliting the dataset into chrun and non-churn price by price date
churn_grp_price = price_data[price_data['churn']==1].groupby(['price_date'])[numerical_features].mean().reset_index()
non_churn_grp_price = price_data[price_data['churn']==0].groupby('price_date')[numerical_features].mean().reset_index()
```

```python
price_features = [val for val in churn_grp_price.columns if churn_grp_price[val].dtypes == 'float64' and val not in['churn']]
price_features
```

```python
for feature in price_features:
    sns.lineplot(churn_grp_price[feature])
    sns.lineplot(non_churn_grp_price[feature])
    plt.xlabel(feature)
    plt.ylabel('churn')
    plt.title(feature)
    plt.legend(['Not Churn','Churn'])
    plt.show()
```

* The analysis suggests that there is little difference in the average price values between churn and non-churn clients. 
* Therefore, based on the available data, it is unlikely that price is a significant factor in determining churn.
