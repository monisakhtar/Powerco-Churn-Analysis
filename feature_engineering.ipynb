## 3. Feature engineering

### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
```

## Transforming Numerical Features

```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'object' and feature!='Id']

## 1.Discrete Variables
discrete_feature=[feature for feature in numerical_features if len(df[feature].unique())<25 and feature!='Id']
    
## 2.Continuous variable
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+['Id']]
   
```

```python
numerical_features
```

```python
discrete_feature
```

```python
continuous_feature
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    sns.histplot(x = feature, data = df)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

* We know from EDA, these continuous features are highly right skewed.
* Performing log transformation on these features would make them more normal distributed.
* It also decreases the effect of any outliers in the data. 
* Not all features require log transformation

```python
# We will be using logarithmic transformation


for feature in continuous_feature:
    data_client=df.copy()
    data_client[feature]=data_client[feature].apply(lambda x:np.log(1+x))
    sns.histplot(x = feature, data = data_client)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

* Many of these features had a 0 value so we needed to add 1 to value instead of normal log transformation.
* We can notice the features are more normally distributed now.

```python
# Transforming continuous features

log_features = [
    'cons_12m', 
    'cons_gas_12m', 
    'cons_last_month',
    'forecast_cons_12m', 
    'forecast_cons_year', 
    'forecast_discount_energy',
    'forecast_meter_rent_12m', 
    'forecast_price_energy_off_peak',
    'forecast_price_energy_peak', 
    'forecast_price_pow_off_peak'
]

for feature in log_features:
    df[feature]=df[feature].apply(lambda x:np.log(1+x))
```

```python
## Lets Find the realtionship between discrete variables and churn

for feature in discrete_feature:
    data_client=df.copy()
    sns.countplot(x=feature, hue='churn', data=data_client)
    plt.xlabel(feature)
    plt.ylabel('churn')
    plt.title(feature)
    plt.show()
```

* Since random forests are less sensitive to feature scaling than certain other machine learning methods, it is not always required to employ StandardScaler or MinMaxScaler for random forest models.

## Transforming Categorical Features

```python
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object' and feature!='id']
categorical_features
```

```python
# Let's look at these features one by one

for feature in categorical_features:
    print(f'Column: {feature}')
    print(f'Number of unique values: {df[feature].nunique()}')
    describe_frame = df[[feature]].value_counts().reset_index(name='count')
    print(describe_frame,'\n'+'--'*40)
```

* 'has_gas' column values are t and f, we can change them to 1, 0 , which would make more sense.
* 'channel_sales' and 'origin_up' has some values with very little occurance compared to other values.
* We can replaces these values with a single type as 'others'

```python
# Transforming 'has_gas' feature - 1 for t and 0 for f

df['has_gas']=df['has_gas'].replace(['t','f'],[1,0])
df['has_gas'].value_counts()
```

```python
# Transforming 'channel_sales' and 'origin_up' features, replacing rarely occuring values with others
val_transform = ['channel_sales', 'origin_up']
for feature in val_transform:
    df.loc[df[feature].value_counts()[df[feature]].values < 10000, feature] = 'others'
```

```python
# Let's look at these features one by one

for feature in val_transform:
    print(f'Column: {feature}')
    print(f'Number of unique values: {df[feature].nunique()}')
    describe_frame = df[[feature]].value_counts().reset_index(name='count')
    print(describe_frame,'\n'+'--'*40)
```

* Even after grouping rare values together, their number is pretty low.

```python
#One hot encoding using get_dummies()
df = pd.get_dummies(df, columns = ['channel_sales', 'origin_up'])
```

## Split the Data

```python
X = df.drop(columns=['id','churn'])
y = df['churn']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training cases: %d\nTest cases: %d' % (X_train.shape[0], X_test.shape[0]))
```

* We split the dataset into training dataset and testing dataset. We will split the data into 70% for training and hold back 30% for testing.

## Handling Imbalnce in the dataset.

We already know from EDA, that there is an imbalnce in the data.

```python
total = len(df['churn'])
ax = sns.countplot(x='churn', data=df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()/2
    y = p.get_y() + p.get_height() + 0.02
    ax.annotate(percentage, (x, y), ha='center')
# set plot title and axis labels
ax.set_title("Churn Distribution")
ax.set_xlabel("Churn")
ax.set_ylabel("Count")
```

```python
X_train.isna().sum()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

```

```python
X_train_new = pd.concat([X_train, X_train_resampled])
y_train_new = pd.concat([y_train, y_train_resampled])
```

To avoid overfitting and to ensure that our model is well-trained on both the original and synthetic data, we combine the original training set with the resampled data to create a new balanced training set.

## Model Fitting - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)
```

```python
# Calculate the classification report

from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
# Plot the confusion matrix
# Define labels for the confusion matrix
labels = ['Not Churn', 'Churn']

# Create DataFrame from the confusion matrix
cm_df = pd.DataFrame(confusion_matrix, index=labels, columns=labels)

# Plot the confusion matrix
sns.heatmap(cm_df, annot=True, cmap='Blues', fmt='g')
```

```python
# Calculate true positive, false positive, true negative, and false negative
tn, fp, fn, tp = confusion_matrix.ravel()
print(f'True positive: {tp}\nFalse positive: {fp}\nTrue negative: {tn}\nFalse negative: {fn}')
```

* The model correctly predicted that 4811 customers would churn (True Positive).
* The model incorrectly predicted that only 3 customers would churn when they actually did not (False Positive).
* The model correctly predicted that 47400 customers would not churn (True Negative).
* The model incorrectly predicted that 331 customers would not churn when they actually did (False Negative).

Based on these results, the model seems to be performing well with a high number of true positives and true negatives. However, the number of false positives and false negatives is not negligible, which may lead to some incorrect predictions.

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, roc_curve

print('Accuracy:', accuracy_score(y_test, y_pred))
print("Overall Precision:",precision_score(y_test, y_pred))
print("Overall Recall:",recall_score(y_test, y_pred))
auc = roc_auc_score(y_test,y_prob[:,1])
print('\nAUC: ' + str(auc))

# calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob[:,1])

# plot ROC curve
fig = plt.figure(figsize=(6, 6))
# Plot the diagonal 50% line
plt.plot([0, 1], [0, 1], 'k--')
# Plot the FPR and TPR achieved by our model
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

* Accuracy is a common evaluation metric that measures the overall performance of the model.
* Precision measures the percentage of correctly predicted positive cases out of all predicted positive cases.
* Recall measures the percentage of correctly predicted positive cases out of all actual positive cases.
* AUC (Area Under the Curve) measures the overall performance of the model in distinguishing between positive and negative cases.

Overall, the Random Forest classifier performed very well in predicting customer churn, with a high accuracy, precision, recall and AUC. The model was able to correctly predict the majority of churn cases, while minimizing the false positive rate. The high AUC score indicates that the model is very good at distinguishing between positive and negative cases.
